```python
import glob
from astropy.io import fits
import numpy as np
import matplotlib.pyplot as plt
%matplotlib notebook
```

```python
channels = ["94", "131", "171", "195", "284", "304"]
```

```python
proc_fits_path = "/Volumes/Herschel/data/proc_fits/"
june_fits_path = "/Volumes/Herschel/data/june_fits/"
```

```python
proc_fits_vals, june_fits_vals = dict(), dict()
for channel in channels:
    proc_fits_vals[channel] = []
    for fn in glob.glob(proc_fits_path + channel + "/*.fits"):
        if "normed" not in fn:
            with fits.open(fn) as img:
                proc_fits_vals[channel].append(np.mean(img[0].data))
    june_fits_vals[channel] = []
    for fn in glob.glob(june_fits_path + channel + "/*.fits"):
        if "normed" not in fn:
            with fits.open(fn) as img:
                june_fits_vals[channel].append(np.mean(img[0].data))   
```

```python
channel = "304"
fig, axes = plt.subplots(2, sharex=True, sharey=True)
axes[0].hist(proc_fits_vals[channel])
axes[1].hist(june_fits_vals[channel])
plt.show()
```
